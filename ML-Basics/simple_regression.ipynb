```python
import numpy as np ## linear algebra
import pandas as pd ## data processing, data transformation, and data analysis
from sklearn.linear_model import LinearRegression ## Use to create a linear regression model
import matplotlib.pyplot as plt ## Use to create a scatter plot and a line plot
from sklearn.model_selection import train_test_split ## Use to split the data into training and testing sets
from sklearn.metrics import mean_squared_error, r2_score ,mean_absolute_error, mean_absolute_percentage_error ## Use to evaluate the performance of the model
```

## Gathering the required data from the salary_dataset

```python
df=pd.read_csv("salary_dataset.csv")
```

## Defining the feature(x) and target(y) for my data

```python
x = df.drop(['Salary','Unnamed: 0'],axis=1)
y = df["Salary"]
```

## Spliting the dataset into training and testing part

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

## Model Selection

```python
model = LinearRegression()
```

## Train Linear Regression Model

```python
model=model.fit(x_train,y_train)
```

## Predicting Values from th Model

```python
predictedSalary=model.predict(x_test)
```

```python
predictedSalary
```

```python
errors = mean_squared_error(y_test, predictedSalary)
print("Mean Squared Error:", errors)

r2 = r2_score(y_test, predictedSalary)
print("R-squared:", r2)

mae = mean_absolute_error(y_test, predictedSalary)
print("Mean Absolute Error:", mae)

mape = mean_absolute_percentage_error(y_test, predictedSalary)
print("Mean Absolute Percentage Error:", mape)
```

## Visualinzing Estimated Best fit Line for the Model

```python
plt.scatter(x,y,color='blue')
plt.plot(x,model.predict(x),color='red')
plt.title("Experience vs Salary")
```
