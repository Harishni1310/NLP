```python
import os
import zipfile

# Create folder
os.makedirs("data/movie-dialog-corpus", exist_ok=True)

# Extract zip file
with zipfile.ZipFile("movie-dialog-corpus.zip", "r") as zip_ref:
    zip_ref.extractall("data/movie-dialog-corpus")

print("Extraction completed successfully")
```

```python
import csv
import random

texts = []

with open("data/movie-dialog-corpus/movie_lines.tsv", encoding="utf-8") as i:
    csv_reader = csv.reader(i, delimiter="\t")

    for line in csv_reader:
        texts.append(line[-1])

random.seed(1)
random.shuffle(texts)

texts = texts[:10]

print(texts)
```
