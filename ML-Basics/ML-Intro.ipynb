### Experimenting with random data

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import numpy as np
```

```python
randomdata = np.random.RandomState(1)
x=10*randomdata.rand(50) ## 50 random numbers between 0 and 10
y=2*x-5+randomdata.rand(50) ## y is a linear function of x with some random noise added to it
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(x[:,np.newaxis],y) ## Fit the model to the data
xfit = np.linspace(0,10,1000) ## Create an array of 1000 evenly spaced numbers between 0 and 10
yfit = model.predict(xfit[:,np.newaxis]) ## Use the model to predict the y values for the x values in xfit

plt.scatter(x,y) ## Create a scatter plot of the original data
plt.plot(xfit,yfit,color='green') ## Plot the regression line
```
